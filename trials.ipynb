```python
# File size reducer
from PIL import Image
import os

def resize_image(input_path, output_path, ratio):
    """
    Resizes an image while maintaining its aspect ratio.

    Args:
        input_path (str): Path to the original image.
        output_path (str): Path to save the resized image.
        new_width (int): The desired new width in pixels.
    """
    img = Image.open(input_path)
    original_width, original_height = img.size
    
    # Calculate new height to maintain aspect ratio

    new_height = int(original_height * ratio)
    new_width = int(original_width * ratio)

    # Resize the image
    resized_img = img.resize((new_width, new_height), Image.LANCZOS) 
    # Image.LANCZOS is a high-quality resampling filter

    resized_img.save(output_path, optimize=True)
    print(f"Image resized and saved to {output_path}")

targeted_folder = "/Users/joshualin/Library/Application Support/minecraft/saves/New World/datapacks/Celledit/assets/celledit/textures/painting"
for each_folder in os.listdir(targeted_folder):
    path = f"{targeted_folder}/{each_folder}"

    if not os.path.isdir(path):
        continue
    if each_folder == ".DS_Store":
        continue

    for image in os.listdir(path):
        if image == ".DS_Store":
            continue
        image_path = path + "/" + image
        file_size_bytes = os.path.getsize(image_path)
        if (file_size_bytes / (1024 * 1024)) > 1:
            resize_image(image_path, image_path, 0.5)


```
